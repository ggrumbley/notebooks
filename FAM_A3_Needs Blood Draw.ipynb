#### SQL Query Instructions
---
Run in BioBank Production on Participants table

```sql
SELECT 
    p1.ucsd_mrn,
    p1.firstname,
    p1.lastname,
    p1.dob,
    col.not_collected
FROM participants p1
LEFT JOIN collections col ON col.participant_id = p1.id
WHERE ucsd_mrn NOT LIKE "" AND not_collected = 0
```
